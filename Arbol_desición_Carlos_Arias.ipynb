```python
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
predictors
```

```python
target
```

```python
corr = predictors.corr()
corr
```

```python
sns.heatmap(corr)
plt.show()
```

```python
# esta tabla y grafica nos ayuda a identificar la variables que estan correlacionadas ya sea de manera directa o inversa 
# los puntos negors son las variables inversamente relacionadas y los mas claros los directamente relacionados.
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.20, random_state=10)
```

```python
# se probaron diversos tama√±os del amuestra de prueba y el 20 % fue el que mejor presicion mostro.
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
# Examinamos la matriz de confusi√≥n
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
